```python
accounts = {"doctor": {"chief": {"password": "1234","last_name": "Default"},
                       "0001": {"password": "Password1","first_name": "Drew","last_name": "Gordon","specialty": "Family Medicine","phone_number": "5127814115"}
                      },
    "nurse": {},
    "patient": {}}

lab_catalog = {"CBC": 25, "CMP": 45, "Lipid Panel": 30, "HbA1c": 25}

class Doctor:


#Initializes class instance based on which user log into the system and uses the attributes associated with that user_id key.
    def __init__(self):
        self.current_user = None

    def login(self):
        self.create_dummy_patient()
        print("\n--- Doctor Login ---")
        user_id = input("User ID: ")
        password = input("Password: ")
        if user_id in accounts["doctor"]:
            if accounts["doctor"][user_id]["password"] == password:
                print(f"Login successful. Welcome Dr. {accounts['doctor'][user_id]['last_name']}!")
                self.current_user = user_id
                self.doctor_menu()
                return
        print("Login failed. Try again.")

#Main menu of doctor functions. Tasks 4 and 5 call functions from other modules.
    def doctor_menu(self):
        while True:
            print("\n--- Doctor Menu ---")
            print("1. Create New Account")
            print("2. View Patient Records")
            print("3. Order Labs")
            print("4. Prescribe Medication")
            print("5. Discharge Patient")
            print("6. Logout")
            task = input("Select an option: ")
            if task == "1":
                account_selection = input("Select account type: ")
                if account_selection.lower().strip() == "doctor":
                    self.create_doctor()
                if account_selection.lower().strip() == "nurse":
                    self.create_nurse()
                if account_selection.lower().strip() == "patient":
                    self.create_patient()
            elif task == "2":
                self.view_pt_record()
            elif task == "3":
                self.order_labs()
            elif task == "4":
                precribe_medication()
            elif task == "5":
                discharge_main(self)
            elif task == "6":
                print("Logging out...")
                break
            else:
                print("Invalid choice. Try again.")

#Creates a fake patient dataset for testing.
    def create_dummy_patient(self):
            accounts["patient"]["001"] = {
                "password": "Password1",
                "first_name": "Jane",
                "last_name": "Doe",
                "age": "30",
                "address": "123 Test Dr",
                "admit_date": "Yestahday",
                "billing_items": "gauze, sheets, piss tube",
                "line_item_cost": 1000,
                "total_bill": 100000000000000,
            }

    def create_doctor(self):
        print("\n--- Create New Account ---")
        __user__id = input("New User ID: ")
        __password = input("Password: ")
        first_name = input("First name: ")
        last_name = input("Last name: ")
        specialty = input("Specialty: ")
        phone = input("Phone number: ")
        accounts["doctor"][__user__id] = {
            "password": __password,
            "first_name": first_name,
            "last_name": last_name,
            "specialty": specialty,
            "phone": phone
        }
        print(f"Doctor account '{new_id}' created successfully.")

    def create_nurse(self):
        print("\n--- Create New Account ---")
        new_id = input("New User ID: ")
        new_password = input("Password: ")
        first_name = input("First name: ")
        last_name = input("Last name: ")
        floor_number = input("Floor number: ")
        accounts["nurse"][new_id] = {
            "password": new_password,
            "first_name": first_name,
            "last_name": last_name,
            "floor_number": floor_number
        }
        print(f"Nurse account '{new_id}' created successfully.")

    def create_patient(self):
        print("\n--- Create New Account ---")
        new_id = input("New User ID: ")
        password = input("New User password: ")
        first_name = input("First name: ")
        last_name = input("Last name: ")
        age = input("Patient age: ")
        address = input("Patient address: ")
        admit_date = input("Date of admission: ")
        billing_items = input("Billing items: ")
        line_item_cost = input("Line item cost: ")
        total_bill = input("Total bill $: ")
        accounts["patient"][new_id] = {
            "password": password,
            "first_name": first_name,
            "last_name": last_name,
            "age": age,
            "address": address,
            "admit_date": admit_date,
            "billing_items": billing_items,
            "line_item_cost": line_item_cost,
            "total_bill": total_bill
        }
        print(f"Patient account '{new_id}' created successfully.")

#Uses the dictionary data to print record report and reformats the keys using .replace and .title to create informative and aesthetic text.
    def view_pt_record(self):
        print("\n--- View Patient Record ---")
        pt_id = input("Enter patient ID: ")
    
        if pt_id in accounts["patient"]:
            patient = accounts["patient"][pt_id]
            print(f"\nPatient Record for ID {pt_id}:")
            for key, value in patient.items():
                print(f"{key.replace('_', ' ').title()}: {value}")
        else:
            print("Patient not found.")

#Adds labs and lab_total for cost as new dictionary keys for the patient dictionary if not already present.
    def order_labs(self):
        print("\n--- Order Labs ---")
        pt_id = input("Enter patient ID: ")
    
        if pt_id in accounts["patient"]:
            patient = accounts["patient"][pt_id]
            if "labs" not in patient:
                patient["labs"] = []
            if "lab_total" not in patient:
                patient["lab_total"] = 0
            print("\n--- Lab Menu ---")
            for lab, cost in lab_catalog.items():
                print(f'{lab}: ${cost}')
            lab_select = input("Select Lab: ")
            if lab_select in lab_catalog:
                patient["labs"].append(lab_select)
                patient["lab_total"] += lab_catalog[lab_select]
                print(f"{lab_select} ordered for patient {pt_id}.")
                print(f"New total lab cost: ${patient['lab_total']}")
            else:
                print("Invalid lab selection.")
        else:
            print("Patient ID not found.")


Doctor().login()
```

```python
def prescribe_medication():
    print("\n--- Prescribe Medication ---")
    pt_id = input("Enter patient ID: ")
    if pt_id in accounts["patient"]:
        patient = accounts["patient"][pt_id]
        if 'medications' not in patient:
            patient['medications'] = []
        __medication__name = input("Medication: ")
        dosage = input("Dosage: ")
        route = input("Route: ")
        frequency = input("Frequency: ")
        med = {"Medication": __medication__name, "Dosage": dosage, "Route": route, "Frequency": frequency}
    
        accounts["patient"][pt_id]["medications"].append(med)
        print(f"{__medication__name} has been prescribed for Patient ID {pt_id}.")
```

```python
def discharge_patient(pt_id, doctor_instance):
    if pt_id not in accounts["patient"]:
        print(f"No patient found with ID: {pt_id}")
        return doctor_instance.doctor_menu()

    patient = accounts["patient"][pt_id]
    filename = f"{patient['first_name'].lower()}_{patient['last_name'].lower()}_discharged.txt"

    print("\n--- Discharge Summary ---")
    print(f"Patient ID: {pt_id}")
    print(f"Name: {patient['first_name']} {patient['last_name']}")
    print("Labs:")
    for lab in patient.get("labs", []):
        print(f"  - {lab}")
    print("Medications:")
    for med in patient.get("medications", []):
        print(f"  - {med}")
    print(f"Total Bill: ${patient.get('total_bill', 0)}")
    print("--------------------------")

    with open(filename, 'w') as file:
        file.write(f"Patient ID: {pt_id}\n")
        file.write(f"Name: {patient['first_name']} {patient['last_name']}\n")
        file.write("Labs:\n")
        for lab in patient.get("labs", []):
            file.write(f"  - {lab}\n")
        file.write("Medications:\n")
        for med in patient.get("medications", []):
            file.write(f"  - {med}\n")
        file.write(f"Total Bill: ${patient.get('total_bill', 0)}\n")

    del accounts["patient"][pt_id]
    print(f"Patient {patient['first_name']} {patient['last_name']} has been discharged and their record deleted.\nFile saved as {filename}.")

def exit_system():
    print("Exiting HIS. Bye bye! Keep up the amazing work! :)")
    return

def discharge_main(doctor_instance):
    while True:
        print("\n--- Hospital Information System ---")
        print("1. Discharge Patient")
        print("2. Exit System")
        print("3. Doctor Menu")
        choice = input("Enter your choice: ")

        if choice == "1":
            pt_id = input("Enter patient ID to discharge: ").strip()
            discharge_patient(pt_id, doctor_instance)
        elif choice == "2":
            exit_system()
        elif choice == "3":
            return doctor_instance.doctor_menu()
        else:
            print("Invalid option. Please try again.")
```
