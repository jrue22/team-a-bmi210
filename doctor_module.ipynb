```python
class Doctor:
    accounts = {
        "doctor": {
            "chief": {
                "password": "1234",
                "last_name": "Default"
            },
            "0001": {
                "password": "Password1",
                "first_name": "Drew",
                "last_name": "Gordon",
                "specialty": "Family Medicine",
                "phone_number": "5127814115"
            }
        },
        "nurse": {
            
        },
        "patient": {
            
        }
    }

    lab_catalog = {
    "CBC": 25,
    "CMP": 45,
    "Lipid Panel": 30,
    "HbA1c": 25,
    "TSH": 30,
    "Urinalysis": 15
    }

    def __init__(self):
        self.current_user = None

    def login(self):
        print("\n--- Doctor Login ---")
        user_id = input("User ID: ")
        password = input("Password: ")


        if user_id in Doctor.accounts["doctor"]:
            if Doctor.accounts["doctor"][user_id]["password"] == password:
                print(f"Login successful. Welcome Dr. {Doctor.accounts['doctor'][user_id]['last_name']}!")
                self.current_user = user_id
                self.doctor_menu()
                return
        print("Login failed. Try again.")

    def doctor_menu(self):
        while True:
            print("\n--- Doctor Menu ---")
            print("1. Create New Account")
            print("2. View Patient Records")
            print("3. Order Labs")
            print("4. Prescribe Medication")
            print("5. Discharge Patient")
            print("6. Logout")
            choice = input("Select an option: ")

            if choice == "1":
                account_choice = input("Select account type: ")
                if account_choice.lower().strip() == "doctor":
                    self.create_doctor()
                if account_choice.lower().strip() == "nurse":
                    self.create_nurse()
                if account_choice.lower().strip() == "patient":
                    self.create_patient()
            elif choice == "2":
                self.create_dummy_patient()
                self.view_pt_record()
            elif choice == "3":
                self.create_dummy_patient()
                self.order_labs()
            elif choice == "4":
                print("Prescribe Medication – [Placeholder]")
            elif choice == "5":
                print("Discharge Patient – [Placeholder]")
            elif choice == "6":
                print("Logging out...")
                break
            else:
                print("Invalid choice. Try again.")
  
    def create_dummy_patient(self):
            Doctor.accounts["patient"]["001"] = {
                "password": "Password1",
                "first_name": "Jane",
                "last_name": "Doe",
                "age": "30",
                "address": "123 Test Dr",
                "admit_date": "Yestahday",
                "billing_items": "gauze, sheets, piss tube",
                "line_item_cost": 1000,
                "total_bill": 100000000000000
            }

    def create_doctor(self):
        print("\n--- Create New Account ---")
        new_id = input("New User ID: ")
        new_password = input("Password: ")
        first_name = input("First name: ")
        last_name = input("Last name: ")
        specialty = input("Specialty: ")
        phone = input("Phone number: ")

        Doctor.accounts["doctor"][new_id] = {
            "password": new_password,
            "first_name": first_name,
            "last_name": last_name,
            "specialty": specialty,
            "phone": phone
        }

        print(f"Doctor account '{new_id}' created successfully.")

    def create_nurse(self):
        print("\n--- Create New Account ---")
        new_id = input("New User ID: ")
        new_password = input("Password: ")
        first_name = input("First name: ")
        last_name = input("Last name: ")
        floor_number = input("Floor number: ")

        Doctor.accounts["nurse"][new_id] = {
            "password": new_password,
            "first_name": first_name,
            "last_name": last_name,
            "floor_number": floor_number
        }

        print(f"Nurse account '{new_id}' created successfully.")

    def create_patient(self):
        print("\n--- Create New Account ---")
        new_id = input("New User ID: ")
        password = input("New User password: ")
        first_name = input("First name: ")
        last_name = input("Last name: ")
        age = input("Patient age: ")
        address = input("Patient address: ")
        admit_date = input("Date of admission: ")
        billing_items = input("Billing items: ")
        line_item_cost = input("Line item cost: ")
        total_bill = input("Total bill amount ($): ")

        Doctor.accounts["patient"][new_id] = {
            "password": password,
            "first_name": first_name,
            "last_name": last_name,
            "age": age,
            "address": address,
            "admit_date": admit_date,
            "billing_items": billing_items,
            "line_item_cost": line_item_cost,
            "total_bill": total_bill
        }

        print(f"Patient account '{new_id}' created successfully.")

    def view_pt_record(self):
        print("\n--- View Patient Record ---")
        pt_id = input("Enter patient ID: ")
    
        if pt_id in Doctor.accounts["patient"]:
            patient = Doctor.accounts["patient"][pt_id]
            print("\nPatient Record:")
            print(f"ID: {pt_id}")
            print(f"First Name: {patient.get('first_name', 'N/A')}")
            print(f"Last Name: {patient.get('last_name', 'N/A')}")
            print(f"Age: {patient.get('age', 'N/A')}")
            print(f"Address: {patient.get('address', 'N/A')}")
            print(f"Admit Date: {patient.get('admit_date', 'N/A')}")
            print(f"billing_items: {patient.get('billing_items', 'N/A')}")
            print(f"Line Item Cost: {patient.get('line_item_cost', 'N/A')}")
            print(f"total_bill: {patient.get('total_bill', 'N/A')}")
        else:
            print("Patient not found.")
  
    def order_labs(self):
        print("\n--- Order Labs ---")
        pt_id = input("Enter patient ID: ")
        if pt_id in Doctor.accounts["patient"]:
            patient = Doctor.accounts["patient"][pt_id]
            print("\n--- Lab Menu ---")
            for lab, cost in Doctor.lab_catalog.items():
                print(f'{lab}: ${cost}')
            
                

Doctor().login()
```
